```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\asimg\\emp_rates.csv")
```

```python
df.head()
```

```python
df_filter = df[df["statefips"] == 17]
```

```python
df_filter= df_filter.replace('.','0')
```

```python
df_filter[["emp_combined","emp_combined_inclow", "emp_combined_incmiddle", "emp_combined_inchigh", "emp_combined_inchigh","emp_combined_ss40", "emp_combined_ss60", "emp_combined_ss65", "emp_combined_ss70" ]]= df_filter[["emp_combined","emp_combined_inclow", "emp_combined_incmiddle", "emp_combined_inchigh", "emp_combined_inchigh","emp_combined_ss40", "emp_combined_ss60", "emp_combined_ss65", "emp_combined_ss70"]].apply(pd.to_numeric)
```

```python
df_filter
```

```python
df_month = df_filter.groupby(["year","month","statefips"], as_index=False).agg({"emp_combined": "mean","emp_combined_inclow" : "mean", "emp_combined_incmiddle" : "mean", "emp_combined_inchigh": "mean", "emp_combined_inchigh": "mean","emp_combined_ss40": "mean", "emp_combined_ss60": "mean", "emp_combined_ss65": "mean", "emp_combined_ss70" : "mean" })  
```

```python
df_final=df_month.rename(columns={"emp_combined": "EmplyRate_Combined","emp_combined_inclow": "EmplyRate_LowIncome", "emp_combined_incmiddle":"EmplyRate_MiddleIncome", "emp_combined_inchigh": "EmplyRate_HighIncome","emp_combined_ss40":"Trade_Transport_Utilites", "emp_combined_ss60":"Professional_Business", "emp_combined_ss65":"Education_Health", "emp_combined_ss70":"Leisure_Hospitality"})
```
