```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\asimg\\UIClaims.csv")
```

```python
df.head()
```

```python
df_filter = df[df["statefips"] == 17]
```

```python
#df_filter.dtypes
df_filter= df_filter.replace('.','0')
df_filter[["initclaims_count_pua","initclaims_rate_pua"]]= df_filter[["initclaims_count_pua", "initclaims_rate_pua"]].apply(pd.to_numeric)
#df_filter.to_csv('c:\\users\\asimg\\Documents\\test.csv')
```

```python
#df_month = df_filter.groupby(["year","month","statefips"])["initclaims_count_regular", "initclaims_count_pua" ].sum()
df_month = df_filter.groupby(["year","month","statefips"], as_index=False).agg({"initclaims_count_regular" : "sum", "initclaims_rate_regular": "mean", "initclaims_count_pua": "sum", "initclaims_rate_pua": "mean"})  
```
